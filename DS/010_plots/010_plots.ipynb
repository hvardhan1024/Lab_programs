```python
# Step 1: Import Required Packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Step 2: Load the Pre-processed CSV File
df = pd.read_csv("../009_EDA/mpg_raw.csv")  # Replace with the actual file path if needed
```

```python


# Step 3: Identify Continuous and Categorical Variables
cats = list(df.select_dtypes(include=['object']).columns)  # Categorical Variables
nums = list(df.select_dtypes(exclude=['object']).columns)  # Continuous Variables

print(f'Categorical variables: {cats}')
print(f'Numerical variables: {nums}')
```

```python

# Step 4 (i): Show the Distribution of Continuous Variables using Histograms
plt.figure(figsize=(12, 8))
df[nums].hist(bins=20, figsize=(12, 8), color='skyblue', edgecolor='black')
plt.suptitle("Distribution of Continuous Variables", fontsize=14)
plt.show()
```

```python

# Step 4 (ii): Identify the Relationship Between Two Continuous Variables using Scatter Plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df['horsepower'], y=df['mpg'], hue=df['origin'])
plt.title("Scatter Plot of Horsepower vs MPG")
plt.xlabel("Horsepower")
plt.ylabel("Miles per Gallon (MPG)")
plt.show()
```

```python

# Step 4 (iii): Find and Display the Frequency of Categorical Values using Count Plot
plt.figure(figsize=(8, 6))
sns.countplot(x='origin', data=df, palette="viridis")
plt.title("Count Plot of Origin")
plt.xlabel("Origin")
plt.ylabel("Count")
plt.show()
```

```python

# Step 4 (iv): Apply Point Plots to Display One Continuous and One Categorical Variable
plt.figure(figsize=(8, 6))
sns.pointplot(x='origin', y='mpg', data=df, capsize=0.2)
plt.title("Point Plot of MPG by Origin")
plt.xlabel("Origin")
plt.ylabel("Miles per Gallon (MPG)")
plt.show()
```
